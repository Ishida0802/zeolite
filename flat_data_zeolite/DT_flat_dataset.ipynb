# dataset 1

```python
import pandas as pd
dataset=pd.read_csv('flat_dataset_1.csv', index_col=0, header=0)
x=pd.concat([dataset.iloc[:, 0:5], dataset.iloc[:, 6:8]], axis=1)
y=dataset.iloc[:, 8]
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=36, shuffle=True, stratify=y, random_state=3)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_predict, StratifiedKFold
fold=StratifiedKFold(n_splits=3, shuffle=True, random_state=3)
import numpy as np
max_depths=np.arange(1, 21, 1)
accuracy_in_cv_all=[]
all_depths=[]
from sklearn import metrics
for max_depth in max_depths:
    model=DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=2)
    model.fit(x_train, y_train)
    estimated_y_in_cv=cross_val_predict(model, x_train, y_train, cv=fold)
    accuracy_in_cv_all.append(metrics.accuracy_score(y_train, estimated_y_in_cv))
    all_depths.append(max_depth)
import matplotlib.pyplot as plt
plt.scatter(all_depths, accuracy_in_cv_all)
plt.xlabel('depth')
plt.ylabel('accuracy_score')
plt.show()
optimal_depth=all_depths[accuracy_in_cv_all.index(max(accuracy_in_cv_all))]

model=DecisionTreeClassifier(max_depth=optimal_depth, min_samples_leaf=2)
model.fit(x_train, y_train)
estimated_y_train=pd.DataFrame(model.predict(x_train))
estimated_y_test=pd.DataFrame(model.predict(x_test))
phase=list(set(y))
phase.sort()
confusion_matrix_train=pd.DataFrame(metrics.confusion_matrix(y_train, estimated_y_train), index=phase, columns=phase)
confusion_matrix_test=pd.DataFrame(metrics.confusion_matrix(y_test, estimated_y_test), index=phase, columns=phase)
display(confusion_matrix_train)
display(confusion_matrix_test)
accuracy_train=metrics.accuracy_score(y_train, estimated_y_train)
accuracy_test=metrics.accuracy_score(y_test, estimated_y_test)
print(f"(Accuracy) :  {accuracy_train},  {accuracy_test}")
confusion_matrix_train.to_csv('matrix_train_dt_1.csv')
confusion_matrix_test.to_csv('matrix_test_dt_1.csv')

from sklearn.tree import export_graphviz
from graphviz import Source
with open('DT_flat_dataset_1', 'w') as f:
    export_graphviz(model, out_file=f, feature_names=x.columns, class_names=model.classes_)
dot_file_path='DT_flat_dataset_1'
=Source.from_file(dot_file_path)
.view()
```
