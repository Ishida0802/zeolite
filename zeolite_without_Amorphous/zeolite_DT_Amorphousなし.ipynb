```python
import pandas as pd
dataset=pd.read_csv('zeolite_433.csv', index_col=0, header=0)
x=pd.concat([dataset.iloc[:, 0:5], dataset.iloc[:, 6:8]], axis=1)
y=dataset.iloc[:, 8]
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_predict
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.25, stratify=y, shuffle=True, random_state=4)
fold=StratifiedKFold(n_splits=3, shuffle=True, random_state=1)
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import numpy as np
max_depths=np.arange(1, 21, 1)
all_depths=[]
accuracy_in_cv_all=[]
for max_depth in max_depths:
    model=DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=3)
    estimated_y_in_cv=pd.DataFrame(cross_val_predict(model, x_train, y_train, cv=fold))
    accuracy_in_cv=metrics.accuracy_score(y_train, estimated_y_in_cv)
    accuracy_in_cv_all.append(accuracy_in_cv)
    all_depths.append(max_depth)
import matplotlib.pyplot as plt
plt.rcParams['font.size']=16
plt.scatter(all_depths, accuracy_in_cv_all)
plt.xlabel('max depth')
plt.ylabel('accuracy in cv')
plt.show()
optimal_depth=all_depths[accuracy_in_cv_all.index(max(accuracy_in_cv_all))]
model=DecisionTreeClassifier(max_depth=optimal_depth, min_samples_leaf=1)
model.fit(x_train, y_train)
print(f"最適な深さ: {optimal_depth}")

estimated_y_train=pd.DataFrame(model.predict(x_train))
estimated_y_test=pd.DataFrame(model.predict(x_test))
phase=list(set(y))
phase.sort()
from sklearn import metrics
confusion_matrix_train=pd.DataFrame(metrics.confusion_matrix(y_train, estimated_y_train), index=phase, columns=phase)
print(metrics.accuracy_score(y_train, estimated_y_train))
display(confusion_matrix_train)

confusion_matrix_test=pd.DataFrame(metrics.confusion_matrix(y_test, estimated_y_test), index=phase, columns=phase)
print(metrics.accuracy_score(y_test, estimated_y_test))
display(confusion_matrix_test)

from sklearn.tree import export_graphviz
from graphviz import Source
with open('zeolite_433', 'w') as f:
    export_graphviz(model, out_file=f, feature_names=x.columns, class_names=model.classes_)
dot_file_path='zeolite_433'
決定木=Source.from_file(dot_file_path)
決定木.view()

importance=pd.DataFrame(model.feature_importances_)
importance.index=x.columns
importance.columns=['importance']
importance

# 保存
confusion_matrix_train.to_csv('confusion_matrix_train_dt.csv')
confusion_matrix_train.to_excel('confusion_matrix_train_dt.xlsx')
confusion_matrix_test.to_csv('confusion_matrix_test_dt.csv')
confusion_matrix_test.to_excel('confusion_matrix_test_dt.xlsx')
importance.to_csv('importances_dt.csv')
importance.to_excel('importances_dt.xlsx')
```
