```python
import pandas as pd
accuracy_dt = pd.read_csv('accuracy_dt_0.35.csv', index_col=0) 
accuracy_rf = pd.read_csv('accuracy_rf_0.35.csv', index_col=0)
accuracy_xgb = pd.read_csv('accuracy_xgb_0.35.csv', index_col=0) 
accuracy_lgbm = pd.read_csv('accuracy_lgbm_0.35.csv', index_col=0) 
```

```python
# DT
mean_of_accuracy_dt=accuracy_dt.iloc[:, 1].mean()
std_of_accuracy_dt=accuracy_dt.iloc[:, 1].std()

# RF
mean_of_accuracy_rf=accuracy_rf.iloc[:, 1].mean()
std_of_accuracy_rf=accuracy_rf.iloc[:, 1].std()

# XGB
mean_of_accuracy_xgb=accuracy_xgb.iloc[:, 1].mean()
std_of_accuracy_xgb=accuracy_xgb.iloc[:, 1].std()

# LGBM
mean_of_accuracy_lgbm=accuracy_lgbm.iloc[:, 1].mean()
std_of_accuracy_lgbm=accuracy_lgbm.iloc[:, 1].std()
```

```python
data = {
    'Model': ['DT', 'RF', 'XGB', 'LGBM'],
    'Mean': [mean_of_accuracy_dt, mean_of_accuracy_rf, mean_of_accuracy_xgb, mean_of_accuracy_lgbm],
    'Std': [std_of_accuracy_dt, std_of_accuracy_rf, std_of_accuracy_xgb, std_of_accuracy_lgbm]
}

# DataFrameに変換
summary = pd.DataFrame(data)
summary
```
