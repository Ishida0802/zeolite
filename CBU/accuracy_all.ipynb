```python
import pandas as pd
summary_02=pd.read_csv('summary_0.2.csv')
summary_025=pd.read_csv('summary_0.25.csv')
summary_03=pd.read_csv('summary_0.3.csv')
summary_035=pd.read_csv('summary_0.35.csv')
summary_04=pd.read_csv('summary_0.4.csv')
summary_045=pd.read_csv('summary_0.45.csv')
summary_05=pd.read_csv('summary_0.5.csv')
```

```python
summary_all=pd.concat([summary_02.iloc[:, 2:], summary_025.iloc[:, 2:], summary_03.iloc[:, 2:],
                       summary_035.iloc[:, 2:], summary_04.iloc[:, 2:], summary_045.iloc[:, 2:],
                       summary_05.iloc[:, 2:]], axis=1)
```

```python
summary_all.columns=['mean_0.2', 'std_0.2', 'mean_0.25', 'std_0.25', 
                     'mean_0.3', 'std_0.3', 'mean_0.35', 'std_0.35', 
                     'mean_0.4', 'std_0.4', 'mean_0.45', 'std_0.45', 'mean_0.5', 'std_0.5']
summary_all.index=['DT', 'RF', 'XGB', 'LGBM']
summary_all
```

# DT

```python
import matplotlib.pyplot as plt
plt.rcParams['font.size']=10

x = [0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]

y = [summary_all.iloc[0, 0], summary_all.iloc[0, 2], summary_all.iloc[0, 4],
     summary_all.iloc[0, 6], summary_all.iloc[0, 8], summary_all.iloc[0, 10],
     summary_all.iloc[0, 12]]

yerr = [summary_all.iloc[0, 1], summary_all.iloc[0, 3], summary_all.iloc[0, 5],
        summary_all.iloc[0, 7], summary_all.iloc[0, 9], summary_all.iloc[0, 11],
        summary_all.iloc[0, 13]]

plt.errorbar(x, y, yerr=yerr, capsize=3, elinewidth=1)
plt.scatter(x, y)
plt.scatter(0.25, 0.701, color='red', label='(0.25, 0.923)')
for xi, yi in zip(x, y):
    plt.text(xi, yi + 0.01, f"{yi:.3f}", ha='center', va='bottom')
plt.ylim(0.6, 0.8)
plt.xlabel('ratio of test data')
plt.ylabel('accuracy')
plt.show()
```
